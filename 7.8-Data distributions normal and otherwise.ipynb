```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1. Beta distribution
beta = np.random.beta(0.5, 0.2, 100)

mean = np.mean(beta)
std = np.std(beta)

plt.hist(beta)

plt.axvline(beta.mean(), color='red', linestyle='solid', linewidth=2)
plt.axvline(beta.mean() + beta.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(beta.mean() - beta.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#2. Chi-square distribution 
chi = np.random.chisquare(20,100)

plt.hist(chi)

plt.axvline(chi.mean(), color='red', linestyle='solid', linewidth=2)
plt.axvline(chi.mean() + beta.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(chi.mean() - beta.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#3. Multivariate normal distribution 
mean = [0, 0]
cov =[[1, 0], [0, 100]] 

mnormal = np.random.multivariate_normal(mean, cov, 100)

plt.hist(mnormal)

plt.axvline(mnormal.mean(), color='red', linestyle='solid', linewidth=2)
plt.axvline(mnormal.mean() + mnormal.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(mnormal.mean() - mnormal.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#4. Logistic distribution 
log = np.random.logistic(0.0, 1.0, 100)

plt.hist(log)

plt.axvline(log.mean(), color='red', linestyle='solid', linewidth=2)
plt.axvline(log.mean() + log.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(log.mean() - log.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#5. Log normal distribution 
lognormal = np.random.lognormal(0.0, 1.0, 100)

plt.hist(lognormal)

plt.axvline(lognormal.mean(), color='red', linestyle='solid', linewidth=2)
plt.axvline(lognormal.mean() + lognormal.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(lognormal.mean() - lognormal.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#6. Uniform distribution 
uniform = np.random.uniform(0.2, 0.8, 100)

plt.hist(uniform)

plt.axvline(uniform.mean(), color='red', linestyle='solid', linewidth=2)
plt.axvline(uniform.mean() + uniform.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(uniform.mean() - uniform.std(), color='red', linestyle='dashed', linewidth=2)

plt.show()
```

## Summary

Except for multivariate normal distribution, logistic distribution, and uniform distribution, the descriptive statistics do not provide useful information about the variable as they are not describing central tendency (mean does not fall in the center, and stanard deviation no longer describes how much variance there is). Mean and standard deviation are useful in describing normal distribution where the mean a measue of central tendency. 

```python
x = np.random.normal(5, 0.5, 1000)
y = np.random.normal(10, 1, 1000)
z = x + y

plt.hist(x, color='green', label='x')
plt.hist(y, color='purple', label='y')
plt.hist(z, color='orange', label='x+y')

plt.axvline(x.mean(), color='grey', linestyle='solid', linewidth=2)
plt.axvline(x.mean() + z.std(), color='grey', linestyle='dashed', linewidth=2)
plt.axvline(x.mean() - z.std(), color='grey', linestyle='dashed', linewidth=2)

plt.axvline(y.mean(), color='grey', linestyle='solid', linewidth=2)
plt.axvline(y.mean() + z.std(), color='grey', linestyle='dashed', linewidth=2)
plt.axvline(y.mean() - z.std(), color='grey', linestyle='dashed', linewidth=2)

plt.axvline(z.mean(), color='grey', linestyle='solid', linewidth=2)
plt.axvline(z.mean() + z.std(), color='grey', linestyle='dashed', linewidth=2)
plt.axvline(z.mean() - z.std(), color='grey', linestyle='dashed', linewidth=2)

plt.legend(loc='upper right')

plt.show()
```

The third variable z is also normally distributed. The mean is the sum of the means of x and y. SD is slightly over 1.
